```python
class Books:
    def __init__(self,name,author,pages):
        self.name=name
        self.author=author
        self.pages=pages
    def __str__(self):
        a=1+2
        return f'{self.name},{a}{self.author},{self.pages}'
    def se(self):
        print(self)

book1=Books("gongheguozhiguagn","wanahimin",300)
book2=Books("gongheguozhiguagn","wn",300)
print(book1)
book2.se()
```
